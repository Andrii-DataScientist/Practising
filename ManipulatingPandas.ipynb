**feature engineering** - can be also called like transforming a DataFrame, mutating a DataFrame, in sense of adding a new column, derived from others

```python
homeless['state_pop_m'] = homeless.state_pop / 1000000
homeless.head()
```

```python
homeless['total'] = homeless.individuals + homeless.family_members

homeless['p_individuals'] = homeless.individuals/homeless.total  

homeless.head()
```

```python
homeless["indiv_per_10k"] = 10000 * homeless.individuals / homeless.state_pop 

high_homelessness = homeless[homeless.indiv_per_10k > 20]

high_homelessness_srt = high_homelessness.sort_values('indiv_per_10k',ascending=False)

result = high_homelessness_srt[['state','indiv_per_10k']]

result
```

```python
homeless["total_per_10k"] = 10000 * homeless.total / homeless.state_pop 

high_homelessness = homeless[homeless.total_per_10k > 20]

high_homelessness_srt = high_homelessness.sort_values('total_per_10k',ascending=False)

result = high_homelessness_srt[['state','total_per_10k']]

result
```

```python
def pct_10(column):
    return column.quantile(0.1)

def pct_20(column):
    return column.quantile(0.2)

homeless.state_pop.agg(pct_10)
```

```python
homeless.state_pop.agg([pct_10,pct_20])
```
